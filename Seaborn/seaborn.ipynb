```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing Data

```python
crash_df = sns.load_dataset('car_crashes')
crash_df.head()
```

## Distribution Plots

### Distribution Plot - Univariate Distribution

```python
sns.displot(crash_df['not_distracted'])
```

```python
sns.distplot(crash_df['not_distracted'], kde=False, bins=25)
```

### Joint Plot

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex')
# sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='kde')
```

### KDE Plot

```python
sns.kdeplot(crash_df['alcohol'])
```

### Pair Plot

```python
# Relationship between the numerical features in the entire dataframe
sns.pairplot(crash_df, kind='reg', hue='abbrev')
```

```python
tips_df = sns.load_dataset('tips')
# 'hue' allow us to visualise on the basis of distribution of a categorical variable
sns.pairplot(tips_df, hue='sex', palette='Blues')
```

### Rug Plot

```python
# Single column of dataframe as sticks. Gives an estimation of density; kind of a dense histogram
sns.rugplot(tips_df['tip'])
```

### Styling

```python
### Setting the graph style
sns.set_style('darkgrid')
# sns.set_style('dark')
# sns.set_style('white')
# sns.set_style('whitegrid')
# sns.set_style('ticks')

### Setting the font style
sns.set_context('paper', font_scale=1.5)        # generally what is used in Jupyter Notebook; decent size
# sns.set_context('talk', font_scale=1.4)       # generally used in presentation; bigger size
# sns.set_context('poster', font_scale=1.4)     # generally used on posters; huge size

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
sns.despine(left=True, bottom=True)             # removes the axis lines
```

## Categorical Plots

### Bar Plot

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.median)
# The vertical line on top refers to the variance in the sets
# Estimator is the way in which the data is aggregated
# In count plot, this estimator is simply the count
```

### Count Plot

```python
sns.countplot(x='sex', data=tips_df)
```

### Box Plot

```python
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')
# The line in the middle is the median
# The box extends for one standard deviation from median
# Whiskers extend to all of the other data aside from those considered outliers

plt.legend(loc=0)
```

### Violin Plot

```python
# It is a combination of box plot and kde plot
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split=True)
# It uses the kde distribution and combines the utility of box plot and kde
```

### Strip Plot

```python
plt.figure(figsize=(8,5))
# Scatter Plot with one variable as categorical
sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True, hue='sex', dodge=True)
# dodge is used to separate the hues
# jitter makes the distribution somewhat more spread-out
```

### Swarm Plot

```python
# Combination of violin and strip plot
sns.violinplot(x='day', y='total_bill', data=tips_df)
sns.swarmplot(x='day', y='total_bill', data=tips_df, color='white')
```

### Palettes

```python
# Gives different colormaps
plt.figure(figsize=(8,5))
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', palette='magma')
```

```python
# Gives different colormaps
plt.figure(figsize=(8,5))
sns.set_style('white')
sns.set_context('talk')
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', palette='afmhot')
plt.legend(loc=0)       # best: 0, upper right:1, upper left:2, lower left: 3, lower right: 4
```

## Matrix Plot

### Heatmaps

```python
# The columns and row indexes should be same
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.5)
crash_mx = crash_df.corr()          # tells how influential on other features
sns.heatmap(crash_mx, annot=True, cmap='Blues')
```

```python
flights = sns.load_dataset('flights')
flights = flights.pivot_table(index='month', columns='year', values='passengers')
# print(flights)
sns.heatmap(flights, cmap='Blues', linecolor='white', linewidths=1)
```

### Cluster Map

```python
# Distance between the points is to be calculated and the closest are joined; this is continued till we join everything
iris = sns.load_dataset('iris')
species = iris.pop('species')       # removes the column from the dataframe and puts that column in another variable
sns.clustermap(iris)
```

```python
sns.clustermap(flights, cmap='Blues', standard_scale=1)
# Here, the data labels are re-positioned to cluster them together
```

### Pair Grid

```python
# To get specific plots from the pair plots
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue='species')
# iris_g.map(plt.scatter)
iris_g.map_diag(plt.hist)
iris_g.map_offdiag(plt.scatter)
```

```python
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
```

```python
iris_g = sns.PairGrid(iris, hue='species', x_vars=['sepal_length', 'sepal_width'], y_vars=['petal_length', 'petal_width'])
iris_g.map(plt.scatter)
iris_g.add_legend()
```

## Facet Grid

```python
tips_fg = sns.FacetGrid(tips_df, col='time', row='smoker')
# tips_fg.map(plt.hist, 'total_bill', bins=8)
tips_fg.map(plt.scatter, 'total_bill', 'tip')
```

```python
tips_fg = sns.FacetGrid(tips_df, col='time', hue='smoker', height=4, aspect=1.3, col_order=['Dinner', 'Lunch'], palette='Set1')
tips_fg.map(plt.scatter, 'total_bill', 'tip', edgecolor='w')
plt.legend()
```

```python
kws = dict(s=50, linewidth=0.5, edgecolor='w')
tips_fg = sns.FacetGrid(tips_df, col='sex', hue='smoker', height=4, aspect=1.3, 
                        hue_order=['Yes', 'No'], hue_kws=dict(marker=['^','v']))
tips_fg.map(plt.scatter, 'total_bill', 'tip', **kws)
```

```python
attention_df = sns.load_dataset('attention')
attention_fg = sns.FacetGrid(attention_df, col='subject', col_wrap=5, height=1.5)
attention_fg.map(plt.plot, 'solutions', 'score', marker='.')
```

## Regression Plots

```python
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)
sns.lmplot(x='total_bill', y='tip', hue='sex', data=tips_df, markers=['o', '^'],
           scatter_kws={'s': 100, 'linewidth': 0.5, 'edgecolor': 'w'})
```

```python
sns.lmplot(x='total_bill', y='tip', col='sex', row='time', data=tips_df)
```
